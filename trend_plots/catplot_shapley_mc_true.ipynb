```python
import seaborn as sns
import pandas as pd
```

```python
true = pd.read_csv("True_Shapley.csv")
g = sns.barplot(x="episodes n", y="average time taken (min)",
                data=true)
```

```python
mc = pd.read_csv("MC_Shapley.csv")
g = sns.barplot(x="episodes M", y="average time taken (min)",
                data=mc)
```

```python
replace = pd.read_csv("Replace_Shapley.csv")
g = sns.lineplot(x="episodes M", y="% deviation of MC-approximation from true shapley",
                data=replace)
```

```python
random = pd.read_csv("Random_Shapley.csv")
g = sns.lineplot(x="episodes M", y="% deviation of MC-approximation from true shapley",
                data=random)
```
